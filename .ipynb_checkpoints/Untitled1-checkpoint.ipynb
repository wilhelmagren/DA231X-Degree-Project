```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xx1 = np.linspace(0, 20, num=40)
xx2 = np.linspace(0, 20, num=40)
xx3 = np.linspace(0, 20, num=40)

f1 = np.sin(xx1)
f2 = np.cos(xx2)
f3 = np.sin(xx3) + 1
```

```python
plt.style.use('seaborn')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
plt.plot(f1, marker='v', ms=5, label='sin')
plt.plot(f2, marker='d', ms=5, label='cos')
plt.plot(f3, marker='s', ms=5, label='sin+1', ls=':')
plt.legend()
```
